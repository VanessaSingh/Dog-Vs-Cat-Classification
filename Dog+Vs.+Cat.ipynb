```python
import pandas as pd
import sys
sys.path.append('/usr/local/lib/python2.7/site-packages')
#import cv2
import glob
import os
import os
from PIL import Image
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
```

```python
path = '/home/vanessa/ML/train'
# Store the image file names in a list as long as they are jpgs
images = [f for f in os.listdir(path) if os.path.splitext(f)[-1] == '.jpg']
train_label = []
train_data = []

for image in images[1:500]:
    print(image)
    img = Image.open('/home/vanessa/ML/train/' + image)
    if(image[:3] == 'dog'):
        train_label.append(1)
    else:
        train_label.append(0)
    img = img.resize((256,256), Image.ANTIALIAS)
    train_data.append(np.asarray(img))
```

```python
train_data = np.array(train_data, dtype=np.uint8)
train_label = np.array(train_label, dtype=np.uint8)
#train_data = train_data.transpose((0, 3, 1, 2))
train_data = train_data.astype('float32')
train_data = train_data/255
train_label = np_utils.to_categorical(train_label, 2)
```

```python
path = '/home/vanessa/ML/test'
# Store the image file names in a list as long as they are jpgs
images = [f for f in os.listdir(path) if os.path.splitext(f)[-1] == '.jpg']
test_label = []
test_data = []

for image in images[1:500]:
    print(image)
    img = Image.open('/home/vanessa/ML/test/' + image)
    img = img.resize((256,256), Image.ANTIALIAS)
    test_data.append(np.asarray(img))
```

```python
test_data = np.array(test_data, dtype=np.uint8)
#test_data = test_data.transpose((0, 3, 1, 2))
test_data = test_data.astype('float32')
test_data = test_data/255
```

```python
train_data = train_data.reshape(train_data.shape[0], 3, 256, 256)
```

```python
test_data = test_data.reshape(test_data.shape[0], 3, 256, 256)
```

```python
from keras.layers import Flatten
```

```python
def myModel():
    model = Sequential()
    model.add(Conv2D(32,(3,3), input_shape=(3,256,256), activation='relu', dim_ordering='tf'))
    #32 no. of filters, 5X5 convolution size, 1 is depth of image
#     model.add(MaxPooling2D(pool_size=(2,2), dim_ordering='tf'))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(200, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
    return model
  
    
    
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data, train_label, test_size=0.050, random_state=42)
```

```python
model = myModel()
model.fit(X_train, y_train, epochs = 25, batch_size=300, verbose=2)
scores = model.evaluate(X_test, y_test, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores[1]*100))
```

```python
predictions = model.predict_classes(test_data, verbose = 0)
submissions=pd.DataFrame({"ImageId": list(range(1,len(predictions)+1)),
                         "Label": predictions})
submissions.to_csv("/home/vanessa/pred.csv", index=False, header=True)
```
